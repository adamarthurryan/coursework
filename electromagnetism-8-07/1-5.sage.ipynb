```python
# calculate the laplacian of f
g(r,a)=1/r^2*(derivative(r^2*derivative(f(r,a),r),r))

# the hand-computed value
h(r,a)=3/(4*pi)*(a^2)/(r^2+a^2)^(5/2)

display(h)
display(bool(g==h))
```
