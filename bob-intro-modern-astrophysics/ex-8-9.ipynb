In part (a) we found 
$\left( N_{II}/N_I \right) ^2 + \left( N_{II}/N_I \right) \gamma - \gamma $

where $\gamma \equiv \left( \frac{m_p}{\rho} \right) \left( \frac{2\pi m_e k T}{h^2} \right)^{3/2} e^{-\chi_1/kT}$

From which the quadratic equation yields $N_{II}/N_I = -\gamma \pm \frac{\sqrt{\gamma^2 +4\gamma}}{2}$


```python
import astropy.constants as const
import astropy.units as u
import numpy as np
import matplotlib.pyplot as plt

def gamma_fn(T, rho, chi):
    return (const.m_p/rho)*(2*np.pi*const.m_e/const.h**2)**(3/2)*(const.k_B*T)**(3/2) * np.exp(chi/const.k_B/T)

T = np.linspace(5000, 25000, 100)*u.K
gamma = gamma_fn(T, 1e-6*u.kg/u.m**3, -13.6*u.eV).decompose()
answer = (-gamma + np.sqrt(gamma**2 + 4*gamma))/2


plt.plot(T, answer)
plt.xlabel("Temperature (K)")
plt.ylabel("$N_{II}/N_t$")
```
