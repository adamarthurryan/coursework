```python
%display latex
a=var('a')
assume(a>0)
psi(x) = sqrt(2/a)*sin(2*pi*x/a)
```

```python
integrate(x*psi^2, x, -oo, oo)
```

```python
psi(n,x)=sqrt(2/a)*sin(n*pi*x/a)
display(integrate(x*psi(2,x)*psi(2,x), x, 0, a).full_simplify())
display(integrate(x*psi(3,x)*psi(2,x), x, 0, a))
```

```python
integrate(derivative(psi(3,x),x)*psi(2,x), x, 0, a/2)
```
