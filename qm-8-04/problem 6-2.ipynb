```python
%display latex
Lx=var('L_x')
Ly=var('L_y')
n=var('n')
l=var('l')

assume(Lx>0)
assume(Ly>0)
assume(n>0)
assume(l>0)
assume(n, 'integer')
assume(l, 'integer')

X(x) = sin(pi*n*x/Lx)
Y(y) = sin(pi*l*y/Ly)

display(X,Y)
```

```python
integrate(X^2, x, 0, Lx)
```

```python
integrate(integrate(X^2*Y^2, x, 0, Lx), y, 0, Ly)
```
