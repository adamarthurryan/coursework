```python
%display latex

lhs(eta) = -eta * 1/tan(eta)
rhs(eta) = sqrt(4*pi^2+eta^2)
display(lhs)
display(rhs)
```

```python
p1=plot(lhs, xmax=20, ymin=-1,ymax=20)
p2=plot(rhs, xmax=20, ymin=-1, ymax=20)
show(p1+p2)
```

```python
#by inspection, solutions around 2.5, 5.5, 8.5, 12
r1=find_root(lhs==rhs, 2, 3)
r2=find_root(lhs==rhs, 5, 6)
r3=find_root(lhs==rhs, 8, 9)
r4=find_root(lhs==rhs, 11,12)

display(r1,r2,r3,r4)
```

```python
# valid solutions have eta^2 < z_0^2 = 4*pi^2
# display associated energy levels in units of V_0

E1=r1^2/(4*pi^2).n()
E2=r2^2/(4*pi^2).n()
display(E1,E2)
```
