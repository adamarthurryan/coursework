```python
var('t, y1, y2')

# Define the system of 1st-order ODEs
# eq1: y1' = y2 (i.e., dy/dt = y')
# eq2: y2' = -y1 (i.e., dy'/dt = -y)
de1 = y2
de2 = -y1

# Initial conditions: [t0, y1(t0), y2(t0)]
# Here: t=0, y=1, y'=0
ics = [0, 1, 0]

# Solve using Runge-Kutta 4th order method
# end_points is the final t value, step is the step size
sol = desolve_system_rk4([de1, de2], [y1, y2], ics=ics, ivar=t, end_points=10, step=0.1)

# Format the solution for plotting or numerical inspection
# sol contains pairs of (t, y1, y2)
solution_points = [(p[0], p[1]) for p in sol]
list_plot(solution_points, plotjoined=True, axes_labels=['t', 'y(t)'])
```
