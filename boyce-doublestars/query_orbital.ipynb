```python
from wds_astropy_table import parse_sixth_orbital

from astropy.io import ascii

import numpy as np
import sys

table = parse_sixth_orbital('orb6orbits.txt')
    
```

```python
# column variables
grade = table['grade']
last_date = table['last_date']
semi_major = table['semi_major']
semi_major_units = table['semi_major_units']
pri_mag = table['pri_mag']
sec_mag = table['sec_mag']
#num_obs = table['num_obs']
period_units = table['period_units']
#last_date_o = table['last_date_o']

delta_mag = np.abs(sec_mag-pri_mag)

# filter columns
low_grade = grade == 5
observable = (semi_major >=5) & (semi_major_units == 'a') & (delta_mag <=2) & (sec_mag <= 13)
period = period_units=='y'

old_orbital = last_date < 2000

filtered = table[low_grade & period & old_orbital]

filtered
```
