```python
from wds_astropy_table import parse_wdsweb
from wds_astropy_table import parse_sixth_orbital

from astropy.table import join
from astropy.io import ascii

import numpy as np
import sys

table_wds = parse_wdsweb('wdsweb_summ2.txt')
table_orbital = parse_sixth_orbital('orb6orbits.txt')

table = join(table_wds, table_orbital, ['id','discoverer_normalized'], uniq_col_name='{col_name}{table_name}', table_names=['', '_o'])
```

```python
table
```

```python
# column variables
grade = table['grade']
last_date = table['last_date']
first_date = table['first_date']
last_sep = table['last_sep']
pri_mag = table['pri_mag']
sec_mag = table['sec_mag']
num_obs = table['num_obs']
period_units = table['period_units']
last_date_o = table['last_date_o']
notes = table['notes']

delta_mag = np.abs(sec_mag-pri_mag)

# filter columns
low_grade = grade == 5
observable = (last_sep>=5) & (last_sep<=120) & (delta_mag <=2) & (sec_mag <= 12)
#period = period_units=='y'
no_bad_notes = (np.strings.find(notes, 'L')==-1) & (np.strings.find(notes, 'S')==-1) & (np.strings.find(notes, 'U')==-1)


old_orbital = True #last_date_o < last_date-10

filtered = table[low_grade & observable & old_orbital & no_bad_notes]

filtered
```

```python
high_grade = (grade >=1) & (grade <= 3)
observable = (last_sep>=5) & (last_sep<=120) #& (delta_mag <=2) & (sec_mag <= 12)
#period = period_units=='y'
no_bad_notes = (np.strings.find(notes, 'L')==-1) & (np.strings.find(notes, 'S')==-1) & (np.strings.find(notes, 'U')==-1)


old_orbital = True #last_date_o < last_date-10

filtered = table[high_grade & observable  ]

filtered
```

```python
# For Enhanced Character-Separated Values (ECSV), which preserves metadata like units and data types:
# ECSV is recommended for reproducible text versions of your table.
filtered.write('wdsweb_filtered.csv', format='ascii.csv', overwrite=True)
```
