Performs some calculations on a given pair of stars with the goal of evaluating whether they are physically associated.

Notes and references from personal communication with Richard Jaworski

1. The rPM (relative proper motion) metric using the method of Harshaw (Harshaw, 2016)
2. Angle between 3d space velocity vectors using the method of Nugent (Nugent, 2025)
3. Probablity that components are within one light year of each other using Buchheim's distance probability calculation method (Sethi et al, 2024)
4. binding energy:
    - luminosity from Gaia or using distance and apparent magnitude
    - masses from luminosity using (Duric, 2004)
    - binding energy from masses using (Letchford et al., 2022)
5. System escape velocity using the method of Bonifacio (Bonifacio et al., 2020)

(citations at bottom)


````python
# first provide the gaia source_id values for the primary and secondary

# AG 342 AB
#gaia_id_pri = 3996306127913903232
#gaia_id_sec = 3996306127914172928

# STF 1645 AB
#gaia_id_pri = 1541667932396172800
#gaia_id_sec = 1541667932396172416

# STF 648 AB
#gaia_id_pri = 181312735893067008
#gaia_id_sec = 181318611408327552

# STF 518 BC
gaia_id_pri=3195919254111315712
gaia_id_sec=3195919254111314816

# DUN 5
#gaia_id_pri=4911306239828325632
#gaia_id_sec=4911306239828325760
````

````python

# now import the data

import numpy as np
import astropy.units as u
from astroquery.gaia import Gaia
from astropy.table import Table
from astropy.table import vstack, unique


query_base = """
SELECT
    TOP 10
    *
    FROM gaiadr3.gaia_source AS g
    JOIN gaiadr3.astrophysical_parameters AS ap ON g.source_id = ap.source_id

    WHERE g.source_id IN ({id_pri}, {id_sec})
"""

query=query_base.format(id_pri=gaia_id_pri, id_sec=gaia_id_sec)
print(query)

job = Gaia.launch_job(query)
results = job.get_results()

results_t = Table()
results_t = vstack([results_t, results])
````

````python
# retrieve the primary and secondary rows
results_t.add_index('source_id')
pri=results_t.loc[gaia_id_pri]
sec=results_t.loc[gaia_id_sec]
display(results_t)
````

````python
# utility functions
import astropy.constants as const
from astropy.coordinates import SkyCoord

# return a SkyCoord object for the given row
def skycoord(row):
    return SkyCoord(ra=row['ra']*u.deg, 
                    dec=row['dec']*u.deg, 
              distance=1000*u.pc/row['parallax'], 
              pm_ra_cosdec=row['pmra']*u.mas/u.year, 
              #pm_ra_cosdec=row['pmra']*np.cos(np.radians(row['dec']))*u.mas/u.year, 
              pm_dec=row['pmdec']*u.mas/u.year, 
              radial_velocity=row['radial_velocity']*u.km/u.s, 
              frame='icrs')
            #obstime='J2016.0')

#magnitude of relative velocity
def vrel_mag(c1, c2):
    def mag(v):
      return np.sqrt(v.d_x**2+v.d_y**2+v.d_z**2)

    v_rel=c1.velocity-c2.velocity
    return mag(v_rel)
````

### 1. Harshaw Metric

returns a metric with the following interpretation:

- metric < 20%: Common Proper Motion pairs (CPM)
- 20% < metric < 60%:  Similar Proper Motion pairs (SPM)
- 60% < metric: Different Proper Motion pairs (DPM)

````python
resultant = np.sqrt((pri['pmdec']-sec['pmdec'])**2 + (pri['pmra']-sec['pmra'])**2)
mag_pri = np.sqrt(pri['pmdec']**2 + pri['pmra']**2)
mag_sec = np.sqrt(sec['pmdec']**2 + sec['pmra']**2)
harshaw = resultant/max(mag_pri, mag_sec)
print(f"{harshaw:.3f}")
````

### 2. Nugent 3D space vectors

returns an angle $θ$ between the proper motion vectors in 3 space

 - $θ < 15°$ are designated as Common Space Motion (CSM) pairs
 - $15° < θ < 30°$ are designated as Similar Space Motion (SSM) pairs
 - $θ > 30°$ are designated as Different Space Motion (DSM) pairs

````python
# estimate stellar proper motion in km/s given observed proper motion and parallax in mas 
def pm_mas_to_kps(pm, parallax):
    return 4.74*pm/parallax

# return a km/s proper motion vector for the given row
def vel_vec(row):
    return (pm_mas_to_kps(row['pmra'], row['parallax']), pm_mas_to_kps(row['pmdec'], row['parallax']), row['radial_velocity'])

# dot product of the given 3 vectors
def dot(v1, v2):
    return v1[0]*v2[0]+v1[1]*v2[1]+v1[2]*v2[2]

# magnitude of the given 3 vector
def mag(v):
    return np.sqrt(v[0]**2+v[1]**2+v[2]**2)


vel_vec_pri = vel_vec(pri)
vel_vec_sec = vel_vec(sec)

print(mag(vel_vec_pri), mag(vel_vec_sec))
print()

# compute the angle between velocity vectors from the dot product
theta = np.arccos(dot(vel_vec_pri, vel_vec_sec)/(mag(vel_vec_pri)*mag(vel_vec_sec)))
print(f"{np.degrees(theta):.3f} degrees")
````

### 4 Binding Energy

$ E = \frac{1}{2} \frac{m_a m_b}{m_a+m_b} v_r^2 - \frac{G m_a m_b}{D} $

The system is bound if E is negative, unbound if E is positive.

````python
c_pri = skycoord(pri)
c_sec = skycoord(sec)
r = c_pri.separation_3d(c_sec)

m_pri = pri['mass_flame']*u.M_sun
m_sec = sec['mass_flame']*u.M_sun

v_rel = vrel_mag(c_pri, c_sec)

e_binding = 1/2 * m_pri*m_sec/(m_pri+m_sec) * v_rel**2 - const.G*m_pri*m_sec/r

print(f"E_binding: {e_binding:.3f}")

if (e_binding>0):
    print("unbound")
else:
    print("bound")
````

### 5 Bonifacio Escape Velocity (with skycoord)

The relative velocity between bound objects should be less than their mutual escape velocity. 
Ie. $v_r < v_e$

$v_e = \sqrt{\frac{2 G (M_a+M_b)}{r}}$ where $r$ is separation distance, $M_a$ and $M_b$ are the stellar masses.

````python
            
#escape velocity for a given system of total mass M, separation r
def escape_v(M, r):
    return np.sqrt(2*const.G*M/r).to(u.km/u.s)

c_pri = skycoord(pri)
c_sec = skycoord(sec)

r = c_pri.separation_3d(c_sec)
M = (pri['mass_flame']+sec['mass_flame'])*u.M_sun

v_escape = escape_v(M,r)

v_rel = vrel_mag(c_pri, c_sec)

print(f"r: {r:.3f}")
print(f"M: {M:.3f}")
print(f"M_pri: {pri['mass_flame']*u.M_sun:.3f}")
print(f"M_sec: {sec['mass_flame']*u.M_sun:.3f}")
print(f'v_escape: {v_escape:.3f}')
print(f'v_rel: {v_rel:.3f}')

if (v_rel<=v_escape):
    print("bound")
else:
    print("unbound")
````
